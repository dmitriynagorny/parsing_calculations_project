```python
import pandas as pd
import plotly.express as px
import re
import pickle

from catboost import CatBoostRegressor
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import GridSearchCV
```

<p color=red>Загрузка и первичная обработка данных:</p>

```python
data_1 = pd.read_csv('/home/dmitriy/PycharmProjects/parsing_project/venv/datas/data_1.csv')
data_2 = pd.read_csv('/home/dmitriy/PycharmProjects/parsing_project/venv/datas/data_2.csv')
data_3 = pd.read_csv('/home/dmitriy/PycharmProjects/parsing_project/venv/datas/data_3.csv')
data_4 = pd.read_csv('/home/dmitriy/PycharmProjects/parsing_project/venv/datas/data_4.csv')

data = pd.concat([data_1, data_2, data_3, data_4], ignore_index=True)
```

```python
data.head(10)
```

```python
def add_columns(data_replace):
    columns = data_replace['Общая информация характеристики']
    specification = data_replace['Общая информация параметры']
    try:
        columns = columns.split('$')
        specification = specification.split('$')
        for count in range(len(columns)):
            data_replace[columns[count]] = specification[count]
    except Exception:
        pass

    return data_replace
```

```python
data_copy = data.apply(lambda x: add_columns(x), axis=1).drop(['Общая информация характеристики', 'Общая информация параметры'], axis=1)
```

```python
data_copy.head(10)
```

```python
data_copy.info()
```

<p color=red>Переведем информацию о времени до метро в численный вид (считаем, что на транспорте в 4 раза дольше):</p>

```python
def categories_metro(datas):
    try:
        times = int(re.findall(r'\b\d+\b', datas)[0])
        if datas.find('пеш') >= 0:
            return times
        else:
            return times * 4
    except Exception:
        return -1
```

```python
data_copy['Время до метро'] = data_copy['Время до метро'].apply(
    lambda x: categories_metro(x)
)
```

<p color=red>Переведем информацию о сроках строительства в численный вид:</p>

```python
def building_time(datas):
    try:
        times = re.findall(r'\b\d+\b', datas)
        if int(times[1]) < 2022:
            return 0
        elif int(times[1]) <= 2023:
            return 1
        else:
            return 2
    except Exception:
        return 0
```

```python
data_copy['Окончание стройки'] = data_copy['Окончание стройки'].apply(
    lambda x: building_time(x)
)
```

<p color=red>Переведем информацию о высоте потолков в численный вид:</p>

```python
def height(datas):
    try:
        heights = re.findall(r'\b\d+\b', datas)
        heights = '.'.join(heights)
        heights = float(heights)
        return heights
    except Exception:
        return float('nan')
```

```python
data_copy['Высота потолков'] = data_copy['Высота потолков'].apply(
    lambda x: height(x)
)
```

<p color=red>Переведем информацию о количестве балконов в численный вид:</p>

```python
def balcony(datas):
    try:
        balconies = re.findall(r'\b\d+\b', datas)
        if len(balconies) == 1:
            return int(balconies[0])
        else:
            sum_balconies = 0
            for count in balconies:
                sum_balconies += int(count)
            return sum_balconies
    except Exception:
        return float('nan')
```

```python
data_copy['Балкон/лоджия'] = data_copy['Балкон/лоджия'].apply(
    lambda x: balcony(x)
)
```

<p color=red>Переведем информацию о количестве уборных в численный вид:</p>

```python
def bathroom(datas):
    try:
        bathrooms = re.findall(r'\b\d+\b', datas)
        if len(bathrooms) == 1:
            return int(bathrooms[0])
        else:
            sum_bathrooms = 0
            for count in bathrooms:
                sum_bathrooms += int(count)
            return sum_bathrooms
    except Exception:
        return float('nan')
```

```python
data_copy['Санузел'] = data_copy['Санузел'].apply(
    lambda x: bathroom(x)
)
```

<p color=red>Обработаем выбросы:</p>

```python
columns = ['Цена за метр', 'Высота потолков', 'Время до метро', 'Площадь', 'Цена']
```

```python
for i in columns:
    fig = px.box(data_copy, y=i)
    fig.show()
```

```python
data_copy = data_copy.loc[(data_copy['Цена за метр'] <= 1500000)]
data_copy = data_copy.loc[(data_copy['Высота потолков'] <= 10)]
data_copy = data_copy.loc[(data_copy['Время до метро'] <= 100)]
data_copy = data_copy.loc[(data_copy['Площадь'] <= 250)]
data_copy = data_copy.loc[(data_copy['Цена'] <= 210000000)]
```

<p color=red>Удалим ненужные столбцы:</p>

```python
data_true = data_copy.drop(['Адресс', 'Цена за метр', 'Url', 'Дата', 'Площадь комнат+ обозначение смежных комнат- обозначение изолированных комнат'], axis=1)
```

```python
data_true.head(112)
```

<p color=red>Заполним пропуски:</p>

```python
data_true['Метро'] = data_true['Метро'].fillna('-1')
data_true['Вид из окон'] = data_true['Вид из окон'].fillna('-1')
data_true['Ремонт'] = data_true['Ремонт'].fillna('-1')
data_true['Тип жилья'] = data_true['Тип жилья'].fillna('-1')
data_true['Отделка'] = data_true['Отделка'].fillna('-1')
data_true['Планировка'] = data_true['Планировка'].fillna('-1')
data_true['Санузел'] = data_true['Санузел'].fillna(1)
data_true['Балкон/лоджия'] = data_true['Балкон/лоджия'].fillna(0)
data_true['Высота потолков'] = data_true['Высота потолков'].fillna(2.6)
data_true['Отделка'] = data_true['Отделка'].fillna('Ремонт')

columns = ['Метро', 'Вид из окон', 'Ремонт', 'Тип жилья', 'Отделка', 'Планировка']
encoder = OrdinalEncoder()
data_true[columns] = encoder.fit_transform(data_true[columns])
```

```python
data_true.head()
```

<p color=red>Поделем данные:</p>

```python
target = data_true['Цена']
features = data_true.drop(['Цена'], axis=1)
```

<p color=red>Создание модели:</p>

```python
# Параметры
parametrs_cat = { 'max_depth': range (6, 10, 1),
                'n_estimators': range (240, 270, 5),
                  'learning_rate': [0.1, 0.2, 0.3]
                  }
model_cat = CatBoostRegressor(random_state=0, task_type="CPU", devices='0:1', logging_level='Silent')
```

```python
%%time
grid_cat = GridSearchCV(model_cat, parametrs_cat, scoring='neg_mean_squared_error', n_jobs=-1, cv=3)
grid_cat.fit(features, target)
grid_cat_list = grid_cat.best_estimator_
grid_cat_score = grid_cat.best_score_
```

```python
print(f'Лучший результат на валидации {abs(grid_cat_score) ** 0.5}')
```

```python
print(f'Лучщии параметры: {grid_cat.best_params_}')
```

<p color=red>Обучимся повторно на лучших параметрах:</p>

```python
model_cat_f = CatBoostRegressor(random_state=0, max_depth=6, n_estimators=260, learning_rate=0.2, verbose=100)
model_cat_f.fit(features, target)
```

<p color=red>Сохраним модель и энкодер:</p>

```python
filename = 'finalized_model.sav'
pickle.dump(model_cat_f, open(filename, 'wb'))

filename = 'encoder_model.sav'
pickle.dump(encoder, open(filename, 'wb'))
```
